```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, cohen_kappa_score
from sklearn.preprocessing import label_binarize

# Define class names
class_names = ["sz", "lpd", "gpd", "oth", "lrda", "grda"]

# Define the confusion matrix for each model (normalized values here, but we'll treat them as example counts)
cm_2dcnn = np.array([
    [0.91, 0.01, 0.01, 0.04, 0.01, 0.02],
    [0.01, 0.96, 0.00, 0.01, 0.01, 0.01],
    [0.01, 0.01, 0.95, 0.02, 0.01, 0.02],
    [0.03, 0.08, 0.02, 0.81, 0.04, 0.03],
    [0.01, 0.01, 0.01, 0.03, 0.95, 0.00],
    [0.03, 0.01, 0.01, 0.04, 0.02, 0.89]
])

cm_htcnn = np.array([
    [0.93, 0.01, 0.01, 0.02, 0.01, 0.02],
    [0.01, 0.97, 0.00, 0.00, 0.01, 0.01],
    [0.01, 0.01, 0.96, 0.01, 0.01, 0.01],
    [0.02, 0.06, 0.02, 0.85, 0.03, 0.02],
    [0.01, 0.01, 0.01, 0.02, 0.96, 0.00],
    [0.02, 0.01, 0.01, 0.02, 0.02, 0.92]
])

# Convert normalized confusion matrices to sample counts (assuming 1000 samples for illustration)
sample_count = 1000
cm_2dcnn_counts = (cm_2dcnn * sample_count).astype(int)
cm_htcnn_counts = (cm_htcnn * sample_count).astype(int)

# Get true labels and predicted labels for each model
y_true_2dcnn, y_pred_2dcnn = [], []
y_true_htcnn, y_pred_htcnn = [], []

# For 2D-CNN
for i, row in enumerate(cm_2dcnn_counts):
    y_true_2dcnn.extend([i] * sum(row))  # Append true class
    for j, count in enumerate(row):
        y_pred_2dcnn.extend([j] * count)  # Append predicted class

# For HT-CNN
for i, row in enumerate(cm_htcnn_counts):
    y_true_htcnn.extend([i] * sum(row))
    for j, count in enumerate(row):
        y_pred_htcnn.extend([j] * count)

# Compute metrics
def compute_metrics(y_true, y_pred, class_names):
    metrics = {}
    metrics['Accuracy'] = accuracy_score(y_true, y_pred)
    metrics['Precision'] = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    metrics['Recall'] = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    metrics['F1 Score'] = f1_score(y_true, y_pred, average='weighted', zero_division=0)
    metrics['Cohen\'s Kappa'] = cohen_kappa_score(y_true, y_pred)

    # Compute AUROC
    y_true_bin = label_binarize(y_true, classes=np.arange(len(class_names)))
    y_pred_bin = label_binarize(y_pred, classes=np.arange(len(class_names)))
    metrics['AUROC'] = roc_auc_score(y_true_bin, y_pred_bin, average='weighted', multi_class='ovr')

    return metrics

# Compute metrics for 2D-CNN
metrics_2dcnn = compute_metrics(y_true_2dcnn, y_pred_2dcnn, class_names)
print("Metrics for 2D-CNN Model (with CE Loss):")
for metric, value in metrics_2dcnn.items():
    print(f"{metric}: {value:.4f}")

# Compute metrics for HT-CNN
metrics_htcnn = compute_metrics(y_true_htcnn, y_pred_htcnn, class_names)
print("\nMetrics for HT-CNN Model (with CE Loss):")
for metric, value in metrics_htcnn.items():
    print(f"{metric}: {value:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# Class names
class_names = ["sz", "lpd", "gpd", "lrda", "grda", "oth"]

# Number of classes
n_classes = len(class_names)

# Define the normalized confusion matrix for each model and convert them to sample counts (assuming 1000 samples)
sample_count = 1000
cm_2dcnn = np.array([
    [0.91, 0.01, 0.01, 0.04, 0.01, 0.02],
    [0.01, 0.96, 0.00, 0.01, 0.01, 0.01],
    [0.01, 0.01, 0.95, 0.02, 0.01, 0.02],
    [0.03, 0.08, 0.02, 0.81, 0.04, 0.03],
    [0.01, 0.01, 0.01, 0.03, 0.95, 0.00],
    [0.03, 0.01, 0.01, 0.04, 0.02, 0.89]
]) * sample_count

cm_htcnn = np.array([
    [0.93, 0.01, 0.01, 0.02, 0.01, 0.02],
    [0.01, 0.97, 0.00, 0.00, 0.01, 0.01],
    [0.01, 0.01, 0.96, 0.01, 0.01, 0.01],
    [0.02, 0.06, 0.02, 0.85, 0.03, 0.02],
    [0.01, 0.01, 0.01, 0.02, 0.96, 0.00],
    [0.02, 0.01, 0.01, 0.02, 0.02, 0.92]
]) * sample_count

# Helper function to generate true and predicted labels based on confusion matrix
def generate_labels_from_confusion_matrix(cm):
    y_true = []
    y_pred = []
    for i, row in enumerate(cm):
        y_true.extend([i] * int(sum(row)))
        for j, count in enumerate(row):
            y_pred.extend([j] * int(count))
    return y_true, y_pred

# Generate true and predicted labels for both models
y_true_2dcnn, y_pred_2dcnn = generate_labels_from_confusion_matrix(cm_2dcnn)
y_true_htcnn, y_pred_htcnn = generate_labels_from_confusion_matrix(cm_htcnn)

# Binarize the labels for ROC curve calculation
y_true_2dcnn_bin = label_binarize(y_true_2dcnn, classes=np.arange(n_classes))
y_pred_2dcnn_bin = label_binarize(y_pred_2dcnn, classes=np.arange(n_classes))

y_true_htcnn_bin = label_binarize(y_true_htcnn, classes=np.arange(n_classes))
y_pred_htcnn_bin = label_binarize(y_pred_htcnn, classes=np.arange(n_classes))

# Compute ROC curve and ROC area for each class for both models
fpr_2dcnn = dict()
tpr_2dcnn = dict()
roc_auc_2dcnn = dict()
fpr_htcnn = dict()
tpr_htcnn = dict()
roc_auc_htcnn = dict()

for i in range(n_classes):
    fpr_2dcnn[i], tpr_2dcnn[i], _ = roc_curve(y_true_2dcnn_bin[:, i], y_pred_2dcnn_bin[:, i])
    roc_auc_2dcnn[i] = auc(fpr_2dcnn[i], tpr_2dcnn[i])
    
    fpr_htcnn[i], tpr_htcnn[i], _ = roc_curve(y_true_htcnn_bin[:, i], y_pred_htcnn_bin[:, i])
    roc_auc_htcnn[i] = auc(fpr_htcnn[i], tpr_htcnn[i])

# Plot all ROC curves for both models
plt.figure(figsize=(10, 8))
for i in range(n_classes):
    plt.plot(fpr_2dcnn[i], tpr_2dcnn[i], linestyle='--', label=f'2D-CNN ROC curve (class {class_names[i]}) (area = {roc_auc_2dcnn[i]:.2f})')
    plt.plot(fpr_htcnn[i], tpr_htcnn[i], linestyle='-', label=f'HT-CNN ROC curve (class {class_names[i]}) (area = {roc_auc_htcnn[i]:.2f})')

# Plotting the random chance line
plt.plot([0, 1], [0, 1], 'k--', lw=2, label="Random Guess")

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison between 2D-CNN and HT-CNN Models")
plt.legend(loc="best")
plt.show()
```
